```python
from collections import Counter
import csv
import codecs
import re
import os
```

```python
def ashindi0(bilingual):
    fle = codecs.open(bilingual, 'r', encoding='cp1251', errors='ignore')
    eng = []
    for line in fle:
        find_adj = re.findall('<se lang="en".*?</se>', line)# ищет строчку с английским 
        if find_adj != []:
            eng.append(find_adj[0])# создаёт массив из этих строчек
    return eng
```

```python
def ashindi(bilingual, eng):
    text_as_hind = ''
    text = codecs.open(bilingual, 'r', encoding='cp1251', errors='ignore')
    x = text.read()
    for sent in eng:# англо-русское соответвие ищет
        try:
            eng_rus = re.findall(sent + '\n.*?\n', x)
            if eng_rus != [] and '<se lang="ru"' in eng_rus[0]:
                text_as_hind = text_as_hind + eng_rus[0] + '\t'
        except:
            continue
    return text_as_hind
```
